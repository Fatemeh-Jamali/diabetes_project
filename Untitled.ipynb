```python
import pandas as pd
df=pd.read_csv("diabetes.csv")
df.head()
```

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x="Outcome",data=df)
plt.title("Diabetes Count(0=No,1=Yes)")
plt.show()
```

```python
# Check columns where a zero value could be unrealistic
cols_with_zeros=['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
# Number of zero values ​​in each column
for col in cols_with_zeros:
    print(f"{col} has {df[df[col]==0].shape[0]} zero values")
```

```python
# Replace zeros with median (a more appropriate value for medical data)
for col in cols_with_zeros:
    median=df[col].median()
    df[col]=df[col].replace(0,median)
```

```python
#Feature normalization
from sklearn.preprocessing import StandardScaler
X=df.drop('Outcome',axis=1)
y=df['Outcome']
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
model=RandomForestClassifier(random_state=42)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
#feaure impotance
import numpy as np
importances=model.feature_importances_
feature_names=df.drop('Outcome',axis=1).columns
feat_importances=pd.Series(importances,index=feature_names)
feat_importances=feat_importances.sort_values(ascending=False)
print(feat_importances)
```

```python
plt.figure(figsize=(8,6))
feat_importances.plot(kind='bar',color='teal')
plt.title('Feature Impotance_Diabetes prediction')
plt.ylabel('Importance Score')
plt.xlabel('Features')
plt.grid(True,linestyle='--',alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
from xgboost import XGBClassifier
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
model_xgb=XGBClassifier(eval_metric='logloss',random_state=42)
model_xgb.fit(X_train,y_train)
y_pred=model_xgb.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
plt.figure(figsize=(8,6))
xgb_importances=model_xgb.feature_importances_
feat_importances=pd.Series(xgb_importances,index=feature_names).sort_values(ascending=False)
feat_importances.plot(kind='bar',color='coral')
plt.title('XGBoost Feature Impotances')
plt.ylabel('Importance Score')
plt.xlabel('Features')
plt.tight_layout()
plt.show()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
rf_model=RandomForestClassifier(random_state=42)
xgb_model=XGBClassifier(eval_metric='logloss',random_state=42)
rf_model.fit(X_train,y_train)
xgb_model.fit(X_train,y_train)
```

```python
rf_preds=rf_model.predict(X_test)
xgb_preds=xgb_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
def evaluate_model(y_true,y_pred,name):
    print(f"{name} Metrics")
    print(f"Accuracy:{accuracy_score(y_true,y_pred):.2f}")
    print(f"Precision:{precision_score(y_true,y_pred):.2f}")
    print(f"Recall:{recall_score(y_true,y_pred):.2f}")
    print(f"F1_score:{f1_score(y_true,y_pred):.2f}")
    print("-"*30)
evaluate_model(y_test,rf_preds,"RandomForest")
evaluate_model(y_test,xgb_preds,"XGBoost")
```

```python
metrics=["Accuracy","Precision","Recall","F1_score"]
rf_scores=[
    accuracy_score(y_test,rf_preds),
    precision_score(y_test,rf_preds),
    recall_score(y_test,rf_preds),
    f1_score(y_test,rf_preds)
]
xgb_scores=[
    accuracy_score(y_test,xgb_preds),
    precision_score(y_test,xgb_preds),
    recall_score(y_test,xgb_preds),
    f1_score(y_test,xgb_preds)
]
x=range(len(metrics))
plt.figure(figsize=(8,5))
plt.bar(x,rf_scores,width=0.4,label="Random Forest",align="center")
plt.bar([i+0.4 for i in x],xgb_scores,width=0.4,label="XGBoost",align="center")
plt.xticks([i+0.2 for i in x],metrics)
plt.ylim(0,1)
plt.title("Comparison of Models")
plt.legend()
plt.grid(True,linestyle='--',alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
xgb_scores=cross_val_score(xgb_model,X_scaled,y,cv=5,scoring="f1")
print("XGBoost mean F1:",xgb_scores.mean())
rf_scores=cross_val_score(rf_model,X_scaled,y,cv=5,scoring="f1")
print("Random Forest mean F1:",rf_scores.mean())
```
